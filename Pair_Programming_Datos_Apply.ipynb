1. Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
df2=pd.read_csv ("pair_programming27dic.csv", index_col = 0)
```

```python
df2
```

```python
pd.set_option('display.max_columns', None)
df2
```

```python
def cambiar_comas (columnas):
    try:
        return float (columnas.replace (",", ".") )
    except:
        return np.nan
```

```python
df2["armed forces size"]=df2["armed forces size"].apply(cambiar_comas)
df2["armed forces size"]
```

```python
df2["co2-emissions"]=df2["co2-emissions"].apply(cambiar_comas)
df2["co2-emissions"]
```

```python
df2["density"]=df2["density"].apply(cambiar_comas)
df2["density"]
```

```python
df2["cpi"]=df2["cpi"].apply(cambiar_comas)
df2["cpi"]
```

```python
def cambiar_comas2 (columnas):
    try:
        return float (columnas.replace (",", "") )
    except:
        return np.nan
```

```python
df2["land area"]=df2["land area"].apply(cambiar_comas)
df2["land area"]
```

```python
df2["population"]=df2["population"].apply(cambiar_comas2)
df2["population"]
```

```python
df2["gdp"]=df2["gdp"].apply(cambiar_comas2)
df2["gdp"]
```
