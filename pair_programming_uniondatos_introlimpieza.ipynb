# Ejercicios de limpieza

1. Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
df_merge.drop("country", axis = 1, inplace = True)
df_merge.head(2)
```

2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
- No tengan espacios.
- Estén en minúscula.
- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
nuevas_columnas = {columna: columna.lower().replace("( %)", " ").replace('\n', '').replace(':', '').replace("(km2)", " ").replace("(%)", " ").strip()  for columna in df_merge.columns}
nuevas_columnas
```

```python
df_merge.rename(columns = nuevas_columnas, inplace = True)
df_merge
```

3. La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
df_merge['coordinates']
```

```python
df_merge['coordinates'].str.split(', ', expand=True) 
```

```python
df_merge[['latitud', 'longitud']] = df_merge["coordinates"].str.split(', ', expand=True).get([0,1])
df_merge.head(2)
```

```python
df_merge.drop(columns=['coordinates'], inplace = True)
df_merge
```

4. Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
columnas = ['unemployment rate', 'total tax rate', 'tax revenue', 'population labor force participation', 'out of pocket health expenditure', 'gross tertiary education enrollment', 'gross primary education enrollment', 'forested area', 'cpi change', 'agricultural land']
for columna in columnas:
    df_merge[columna] = df_merge[columna].str.replace('%', '', regex=True)
df_merge
```

5. Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
columna_dolar = ['gasoline price', 'gdp', 'minimum wage']
for columna in columna_dolar:
    df_merge[columna] = df_merge[columna].str.replace('$', '', regex=False)
df_merge
```

6. Guarda el DataFrame para usarlo en el pairprogramming de mañana.

```python
df_merge.to_csv('pair_programming27dic.csv', index=False)
```
